```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import matplotlib.pyplot as plt  # for plotting facilities
```

```python
data = '/content/Wholesale customers data.csv'

df = pd.read_csv(data)
```

```python
X = df.drop('Channel', axis=1)

y = df['Channel']
```

```python
X.head()
```

```python
y.head()
```

```python
y.head().unique()
```

```python
# convert labels into binary values

y[y == 2] = 0

y[y == 1] = 1
```

```python
y
```

```python
# import XGBoost
import xgboost as xgb
# define data_dmatrix
data_dmatrix = xgb.DMatrix(data=X,label=y)
```

```python
data_dmatrix
```

```python
# split X and y into training and testing sets

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 0)
```

```python
# import XGBClassifier
from xgboost import XGBClassifier


# declare parameters
params = {
            'objective':'binary:logistic',
            'max_depth': 4,
            'alpha': 10,
            'learning_rate': 1.0,
            'n_estimators':100
        }

# instantiate the classifier
xgb_clf = XGBClassifier(**params)
# fit the classifier to the training data
xgb_clf.fit(X_train, y_train)
```

```python
print(xgb_clf)
```

```python
# make predictions on test data

y_pred = xgb_clf.predict(X_test)
```

```python
y_pred
```

```python
# compute and print accuracy score

from sklearn.metrics import accuracy_score

print('XGBoost model accuracy score: {0:0.4f}'. format(accuracy_score(y_test, y_pred)))
```

```python
from xgboost import cv

params = {"objective":"binary:logistic",'colsample_bytree': 0.3,'learning_rate': 0.1, 'max_depth': 5, 'alpha': 10}

xgb_cv = cv(dtrain=data_dmatrix, params=params, nfold=3,
                    num_boost_round=50, early_stopping_rounds=10, metrics="auc", as_pandas=True, seed=123)
```

```python
xgb_cv
```

```python
xgb_cv.head()
```

```python
# Get the overall mean AUC score from the test set
overall_test_auc = xgb_cv['test-auc-mean'].max()

# Print the overall score
print(f"Overall test AUC score: {overall_test_auc}")
```
