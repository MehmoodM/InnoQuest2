```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.set(style='whitegrid')
plt.figure(figsize=(10,6))
sns.histplot(data['median_house_value'], kde=True)
plt.title('Distribution of Median House Value')
plt.xlabel('Median House Value')
plt.ylabel('Frequency')
plt.show()
```

```python
sns.set(style='whitegrid')
plt.figure(figsize=(10,6))
sns.histplot(data_cleaned['median_house_value'], kde=True)
plt.title('Distribution of Median House Value')
plt.xlabel('Median House Value')
plt.ylabel('Frequency')
plt.show()
```

```python
# IQR = Q3 - Q!
# lb = Q1 - 1.5 * IQR
# ub = Q3+ 1.5 * IQR

Q1 = data_cleaned['median_house_value'].quantile(0.25)
Q3 = data_cleaned['median_house_value'].quantile(0.75)
print(Q1)
print(Q3)
```

```python
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
```

```python
data_no_outlier = data_cleaned[(data_cleaned['median_house_value'] >= lower_bound) & (data_cleaned['median_house_value'] <= upper_bound)]
data_no_outlier.shape
```

```python
data_no_outlier.head()
```

```python
data_cleaned.shape
```

```python
data_cleaned.shape[0] - data_no_outlier.shape[0]
```

```python
sns.set(style="whitegrid")
plt.figure(figsize=(10, 6))
sns.histplot(data_no_outlier['median_house_value'], kde=True)
plt.title('Distribution plot')
plt.show()
```

```python

sns.set(style="whitegrid")
plt.figure(figsize=(10, 6))
sns.boxplot(data_no_outlier['median_house_value'])
plt.title('Box plot')
plt.show()
```
